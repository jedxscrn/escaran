### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


from

$c_k=\sum_{n=0}^{N-1} y_n \exp \left(-\mathrm{i} \frac{2 \pi k n}{N}\right)$

we can obtain function dft()

````python
import numpy as np
import math
import matplotlib.pyplot as plt

def dft(y):                                                               # function to calculate discrete Fourier transforms coefficients
  N = len(y)
  c = np.zeros(N//2+1,complex)
  for k in range(N//2+1):
    for n in range(N):
      c[k] += y[n]*np.exp(((-2j)*(np.pi)*k*n)/N)
  return c

N = 1000                                                                  # given points
x = np.linspace(0, N, N)                                                  # N evenly spaced points
arr = []                                                                  # array to save results

def yn(x,N):                                                              # modulated sine wave
  for i in range(len(x)):                                                 # iterating for each point in x
    y = math.sin(math.pi*x[i]/(N))*math.sin(20*math.pi*x[i]/(N))              # given
    arr.append(y)
  return arr

y = yn(x,N)
plt.grid(alpha = 0.5)
plt.plot(np.abs(dft(y)), 'ro-')                                           # calculates the independent DFT coefficients of y
plt.xlim(0,50)                                                            # limiting x axis from 500 to 50 to be able to see peak clearly
plt.xlabel("Frequency")
plt.ylabel("Coefficients")
plt.title("Forward DFT of modulated sine wave")
plt.show()

````

we can observe in the plot that it peaks at 10hz, which is consistent with my initial observation from the equation (since the frequency of $\sin(\pi n/N)$ is $0.5\ hz$ and $sin(20\pi n/N)$ is $10\ hz$).

also, since the frequency of $\sin(\pi n/N)$ is smaller than of $sin(20\pi n/N)$, it acts as the "envelope" of the 10hz, which can be seen in the plot as the slope going up and down the peak.

<span style="color:blue"> In fact, since the product of two sines is the sum of two pure sine frequencies, you can show that there are two pure frequencies at 9.5 Hz and 10.5 Hz. Since the DFT does not sample exactly at these frequencies (they are "in-between" the sample points), you observe this enveloope. </span>

Now you can start to write code for this problem!

````python
dft_coefficients = np.fft.rfft(dow)                    # calculating coefficients of the dft of the data using the function rfft from numpy, which returns an array of 1N/2 + 1 complex numbers

ten = int(0.1*len(dft_coefficients))                   # setting all but the first 10% of coefficients in the array to 0 to eliminate noise
dft_coefficients[ten:] = 0                         # if more than 90% is set to 0: inverse Fourier transform plot will be more smoother since more noise is eliminated, but might be less accurate
                                                   # if less than 90% is set to 0: less noise will be filtered and plot will be more similar to original
ift = np.fft.irfft(dft_coefficients)                   # obtaining inverse Fourier transform using the function irfft from numpy

plt.figure(figsize=(10, 6))                            # plotting
plt.plot(dow, label='original data', linewidth=1)
plt.plot(ift, label='inverse Fourier transform', linewidth=2)
plt.title("2006 - 2010 Dow Jones Industrial Average daily closing value")
plt.xlabel("Days")
plt.ylabel("Value")
plt.legend(loc='best')
````

after calculating and plotting the inverse Fourier transform of the coefficients array, with the latter 90% being 0s, we can observe that noise was reduced and the plot is smoother than the original.

# What is happening when you set the Fourier coefficients to zero?
When the Fourier coefficients are set to zero, noise is being eliminated/filtered, resulting in a smoother plot. If >90% of the Fourier coefficients is set to 0, it will be a lot smoother, but may consequently be less accurate relative to the original data. If <90% of the Fourier coefficients is set to 0, plot will be as noisy and similar to the original.
